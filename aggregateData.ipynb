```python
import pandas as pd
```

```python
summary = pd.read_csv('compiledData/compiledSummaryData.csv', index_col=False)
tonnage = pd.read_csv('compiledData/compiledTonnageData.csv', index_col=False)
dimensions = pd.read_csv('compiledData/compiledDimensionsData.csv', index_col=False)
particulars = pd.read_csv('compiledData/compiledParticularsData.csv', index_col=False)
```

```python
dimensions[dimensions['VesselId'] == 1]
```

```python
dimensions = dimensions.astype({'VesselId': 'int64', 'BreadthInFeet': 'float64', 'DepthInFeet': 'float64','LengthInFeet': 'float64' })

lengths = dimensions[dimensions['DimensionTypeLookupId'] == 11]
lengths = lengths[['VesselId', 'LengthInFeet']]
lengths = lengths.drop_duplicates(['VesselId'])
lengths = lengths.set_index('VesselId')

widths = dimensions[dimensions['DimensionTypeLookupId'] == 2]
widths = widths[['VesselId', 'BreadthInFeet']]
widths = widths.drop_duplicates(['VesselId'])
widths = widths.set_index('VesselId')

depths = dimensions[dimensions['DimensionTypeLookupId'] == 2]
depths = depths[['VesselId', 'DepthInFeet']]
depths = depths.drop_duplicates(['VesselId'])
depths = depths.set_index('VesselId')

dims = lengths.join(widths)
dims = dims.join(depths)
```

```python
dims
```

```python
tonnage[tonnage['VesselId'] == 1]
```

```python
tonnage = tonnage.astype({'VesselId': 'int64', 'MeasureOfWeight': 'float64', 'TonnageTypeLookupId': 'int64','TonnageUnitOfMeasurementFilterLookupId': 'int64' })

grt = tonnage[(tonnage['TonnageTypeLookupId'] == 3) & (tonnage['TonnageUnitOfMeasurementFilterLookupId'] == 111)]
grt = grt[['VesselId', 'MeasureOfWeight']]
grt = grt.drop_duplicates(['VesselId'])
grt = grt.set_index('VesselId')
grt = grt.rename(columns={'MeasureOfWeight': 'grt'})

gt = tonnage[(tonnage['TonnageTypeLookupId'] == 1) & (tonnage['TonnageUnitOfMeasurementFilterLookupId'] == 111)]
gt = gt[['VesselId', 'MeasureOfWeight']]
gt = gt.drop_duplicates(['VesselId'])
gt = gt.set_index('VesselId')
gt = gt.rename(columns={'MeasureOfWeight': 'gt'})

tons = grt.join(gt)
```

```python
tons
```

```python
particulars = particulars.astype({'VesselId': 'int64'})
particulars = particulars[['VesselId', 'ServiceSubType', 'IdentificationTypeLookupName', 'CargoAuthorizationDescription']]
particulars = particulars.set_index('VesselId')
```

```python
particulars
```

```python
summary = summary[['VesselId', 'VesselName','VesselCallSign', 'ServiceType', 'ConstructionCompletedYear', 'StatusLookupName', 'Identification', 'ManufacturerHullNumber', 'CountryLookupName', 'HIN', 'OutOfServiceDate']]
summary = summary.set_index('VesselId')
```

```python
summary
```

```python
fullSummary = summary.join(particulars)
fullSummary = fullSummary.join(tons)
fullSummary = fullSummary.join(dims)
```

```python
fullSummary
```

```python
fullSummary = fullSummary[['VesselName', 'VesselCallSign', 'ServiceType', 'ServiceSubType', 
       'ConstructionCompletedYear',  'Identification', 'IdentificationTypeLookupName',
       'grt', 'gt', 'LengthInFeet',
       'BreadthInFeet', 'DepthInFeet', 'CountryLookupName', 'ManufacturerHullNumber', 'HIN', 'StatusLookupName', 'CargoAuthorizationDescription', 'OutOfServiceDate', ]]
```

```python
fullSummary = fullSummary.rename(columns={'IdentificationTypeLookupName': 'IdentificationType', 'CountryLookupName': 'Country'})
```
